```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import numpy as np
import requests
import pprint
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
links = []
page = 1
while True :
            r = requests.get(f'https://genius.com/api/artists/29060/songs?page={page}&sort=popularity')
            if r.status_code == 200:

                response = r.json().get('response')
                next_page = response.get('next_page')
                songs = response.get('songs')

                for song in songs:
                    links.append(song.get('url'))


                page = page+1

                if not next_page :
                    print("plus de nouvelle page")
                    break
```

```python
links
```

```python
def extract_lyrics(url):
    r = requests.get(url)
    
    soup = BeautifulSoup(r.content, 'html.parser')
    lyrics = soup.find('div', class_='lyrics')
    
#      des fois la requete ne marche pas. Je ne sais pas pourquoi. Donc ca ne marche pas il faut
#      relancer la fonction :

    if not lyrics:
        print ('no lyrics')
        return extract_lyrics(url)
    
#     methode de beautifulsoup : .stripped_string permet d'enlever tous les trucs qui ne servent à rien
    lyrics = lyrics.stripped_strings
    parole = []
    for sentence in lyrics:
        parole.append(sentence)
    return parole
    
```

```python
extract_lyrics('https://genius.com/Saez-ptite-pute-lyrics')
```

```python
parole = []
for link in links :
    parole.extend(extract_lyrics(link))
```

```python
# les caracteres uniques des paroles
vocab = sorted(set(parole))
print(vocab)
len(vocab)
```

```python
# on attribue a chaque lettre un chiffre
```

```python
to_number = {u:i for i, u in enumerate(vocab)}
to_char = np.array(vocab)
encoded_parole = []
for x in parole:
    encoded_parole.append(to_number[x])
```

```python
len(encoded_parole) == len(parole)
```

```python
encoded_parole = np.array(encoded_parole)
```

```python
char_dataset = tf.data.Dataset.from_tensor_slices(encoded_parole)
```

```python
sequences = char_dataset.batch(51, drop_remainder=True)
```

```python
def create_seq_targets(seq):
    input_txt = seq[:-1]
    target_txt = seq[1:]
    return input_txt, target_txt
```

```python
dataset = sequences.map(create_seq_targets)
```

```python
dataset = dataset.shuffle(10000).batch(128, drop_remainder=True)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dense,Embedding,Dropout,GRU
```

```python
embed_dim = 64
rnn_neurons = 1026

def create_model(embed_dim, rnn_neurons):
    model = Sequential()
    model.add(Embedding(input_dim = len(vocab), output_dim = embed_dim, batch_input_shape = [128, 50]))
    
    model.add(GRU(units = rnn_neurons ,return_sequences=True,stateful=True,
                  recurrent_initializer='glorot_uniform'))
              
    # Couche Finale Dense de Prédiction
    model.add(Dense(len(vocab), activation = 'softmax'))
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy') 
    return model
```

```python
model = create_model(embed_dim, rnn_neurons)
```

```python
model.summary()
```

```python
model.fit(dataset, epochs = 30)
```
